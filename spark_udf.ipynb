```
nytaxi_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("codec", "gzip") \
    .load("/databricks-datasets/nyctaxi/tripdata/yellow/yellow_tripdata_2009-01.csv.gz")
```

```
nytaxi_df.show()
```

```
# Define a Python function to concatenate two strings
def concat_strings(s1, s2):
    return s1 + "-" + s2
```

```
# Register the function as a UDF
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

concat_udf = udf(concat_strings, StringType())
```

```
df1 = nytaxi_df.withColumn("vendor_payment", concat_udf(nytaxi_df.vendor_name, nytaxi_df.Payment_Type))
```

```
df1.show()
```

```
df1.select("vendor_payment").show()
```

```
df1.select("vendor_payment","Total_Amt").show()
```

```
from pyspark.sql.functions import concat, col

df2 = nytaxi_df.withColumn("vendor_payment", concat(col("vendor_name"), col("Payment_Type")))
df2.select("vendor_payment","Total_Amt").show()
```

```
from pyspark.sql.functions import concat, col, lit

df2 = nytaxi_df.withColumn("vendor_payment", concat(col("vendor_name"), lit("-"), col("Payment_Type")))
df2.select("vendor_payment","Total_Amt").show()
```

```
df2.withColumn("country",lit("USA")).show()
```

```
df3 = df2.select("vendor_payment","Total_Amt").withColumn("country",lit("USA"))
df3.show()
```

```
from pyspark.sql.types import StringType, ArrayType

# Define a UDF that splits a string column into words and returns the words as an array of strings
split_words_udf = udf(lambda text: text.split(), ArrayType(StringType()))
```

```
# Create a DataFrame with a string column
df = spark.createDataFrame([(1, "data engineering"), (2, "delta lake lakehouse")], ["id", "text"])
df.show()
```

```
df = df.withColumn("words", split_words_udf(df["text"]))
```
